```python
def generate_transcript(cname, role, skills):
    text = f"Generate a interview transcript with max 14 to 20 dialogues for a candidate named {cname} applying for the role of {role}. The candidate has skills in {', '.join(skills)}. The transcript should include interviewer's questions, candidate's responses.do not print **"\
           f"do not print any starting lines"
           f"do not print here is a sample interview ..."

    response = client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",  # Specify the model to use
        messages=[{"role": "user", "content": text}]
    )

    transcript = response.choices[0].message.content  # Get the generated transcript
    return transcript

def generate_resume(cname, role, skills):
    text = f"Generate a professional resume for a candidate named {cname} applying for the role of {role}. The candidate's skills are {', '.join(skills)}. do not print **"

    response = client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",  # Specify the model to use
        messages=[{"role": "user", "content": text}]
    )

    resume = response.choices[0].message.content  # Get the generated resume
    return resume
```

```python
def generate_job_description(role):
    """
    Generate a brief job description for the given role using Together AI.
    The job description will consist of 4-5 key points describing the job responsibilities and expectations.
    """
    # Changed: text is now a string instead of a set
    text = f"Generate a brief job description for the role of {role}. The description should contain 4-5 key points describing the job responsibilities and expectations." \
           f"do not print this Here is a brief job description for the role directly give the description"
    response = client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",  # Specify the model to use
        messages=[{"role": "user", "content": text}]
    )

    job_description = response.choices[0].message.content  # Get the generated job description
    return job_description

def generate_reason(result, skills, role, years):
    """
    Generates a reason for selection or rejection based on interview result, skills, role, and years of experience.
    """
    skills_str = ", ".join(skills)  # Convert skills list to a comma-separated string

    if result == "Select":
        # Construct the select reason
        # Changed: text is now a string instead of a set
        text = f"genearte resons of selection refer skills {skills_str} for the job role {role} do not start like 4 reasons for reject directly give the lines. write 3 to 4 lines. do not print **"\
               f"do not strictly print this  Here are some selection reason or Here are the reasons for selection for a role "\
               f"do not print additional lines directly give reasons"
        response = client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",  # Specify the model to use
        messages=[{"role": "user", "content": text}])

        reason = response.choices[0].message.content  # Get the generated job description


    elif result == "Reject":
        # Construct the reject reason
        text = f"genearte resons of rejection refer skills {skills_str} for the job role {role} do not start like 4 reasons for reject directly give the lines. write 3 to 4 lines. do not print **"\
               f"do not strictly print this  Here are the reasons for rejection for a role "\
               f"do not print additional lines directly give reasons"
        response = client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",  # Specify the model to use
        messages=[{"role": "user", "content": text}])

        reason = response.choices[0].message.content  # Get the generated job description

    return reason


def analyze_performance(transcript, resume, skills, role):
    """
    Analyzes the generated transcript and resume to determine the candidate's performance.
    Returns "Select" or "Reject" and a reason based on the interview feedback and resume details.
    """
    # Keywords for analyzing performance
    positive_keywords = ["outstanding", "exceptional", "strong", "excellent", "impressive"]
    neutral_keywords = ["good", "decent", "adequate"]
    negative_keywords = ["lacking", "failed", "needs improvement", "weak"]

    # Check if transcript contains positive, neutral, or negative keywords
    transcript_lower = transcript.lower()
    resume_lower = resume.lower()

    # Determine performance result based on transcript and resume analysis
    if any(keyword in transcript_lower for keyword in positive_keywords):
        result = "Select"
    elif any(keyword in transcript_lower for keyword in negative_keywords):
        result = "Reject"
    else:
        result = "Reject" # Default to "Good" if neither positive nor negative keywords are found

    # Example experience years (this could be parsed from the resume)
    years_of_experience = random.randint(2, 10)  # Example experience years between 2 and 10

    # Generate the reason for the selected candidate
    reason = generate_reason(result, skills, role, years_of_experience)  # Corrected call, removed job_description

    return result, reason
```

```python
id_counter = 1

# Generating interview transcripts and resumes for candidates individually and sequentially
for _ in range(200):  # Example loop for 200 candidates
    start_time = time.time()

    # Generate candidate name
    first_name = random.choice(first_names)
    last_name = random.choice(last_names)
    cname = f"{first_name} {last_name}"

    # Select random job role and skills
    role_data = random.choice(designations)
    role = role_data[0]
    skills = role_data[1]["skills"]

    # 1. Generate the job description for the role
    print(f"Generating job description for {role}...")
    job_description = generate_job_description(role)
    time.sleep(3)  # Sleep between API calls to avoid hitting the rate limit

    # 2. Generate the transcript
    print(f"Generating transcript for {cname} applying for {role}...")
    transcript = generate_transcript(cname, role, skills)
    time.sleep(3)  # Sleep between API calls to avoid hitting the rate limit

    # 3. Generate the resume
    print(f"Generating resume for {cname} applying for {role}...")
    resume = generate_resume(cname, role, skills)
    time.sleep(3)  # Sleep between API calls to avoid hitting the rate limit

    # 4. Analyze performance based on transcript, resume, and job description
    selection_status, reason = analyze_performance(transcript, resume, skills, role)  # Corrected

    # 5. Print the generated reason for selection/rejection
    print(f"Generated reason for {cname}: {reason}")

    # 6. Append data to the list with sequential ID and first name-based ID
    data.append({
        "id": f"{first_name}_{id_counter}",  # ID includes first name and sequential number
        "cname": cname,
        "role": role,
        "transcript": transcript,
        "resume": resume,
        "job_description": job_description,
        "status": selection_status,
        "reason": reason,
    })

    # Increment the sequential ID counter
    id_counter += 1

    end_time = time.time()
    print(f"Transcript {_ + 1} generated in {end_time - start_time:.4f} seconds.")
    time.sleep(60/10)
# Convert to DataFrame
df = pd.DataFrame(data)
print(df)
```
